Quick note book to remove columns from complete measures and add in score

```python
import pandas as pd
```

```python
comp_measures = pd.read_csv("../data/processed/complete_measures.csv")
raw_data = pd.read_csv("../data/processed/historical_scores.csv")
scores = raw_data["FTR"]
comp_measures["scores"] = scores
comp_measures = comp_measures.drop(["Unnamed: 0", "Unnamed: 0.1"], axis=1)
comp_measures.head()
```

```python
comp_measures.set_index("index").to_csv("../data/processed/complete_measures.csv")
```
